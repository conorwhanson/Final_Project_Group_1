```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
pd.set_option('display.max_colwidth', None)
```

```python
emp_survey_df = pd.read_csv("resources/employee_survey_data.csv")
general_df = pd.read_csv("resources/general_data.csv")
manager_df = pd.read_csv("resources/manager_survey_data.csv")
```

```python
emp_survey_df.head()
```

```python
general_df.head()
```

```python
manager_df.head()
```

```python
# Drop the EmployeeCount column as it's all 1's
general_df.drop('EmployeeCount', axis=1, inplace=True)
```

```python
general_df.head()
```

```python
# Create a function to convert rupees to USD
def rupee_to_usd(x):
    return x * .013
```

```python
# Apply rupee_to_usd function to MonthlyIncome column
general_df["MonthlyIncome"] = general_df["MonthlyIncome"].apply(rupee_to_usd)
general_df.head()
```

```python
# create a function to convert km to miles
def km_to_mi(km):
    return km / 1.6
```

```python
general_df["DistanceFromHome"] = general_df["DistanceFromHome"].apply(km_to_mi)
general_df.head()
```

```python
# merge the dataframes
emp_manag_df = pd.merge(emp_survey_df, manager_df, on="EmployeeID")
emp_manag_df.head()
```

```python
# merge emp_manag_df with general_df
hr_df = pd.merge(general_df, emp_manag_df, on="EmployeeID")
hr_df.head()
```

```python
hr_df.info()
```

```python
hr_df = hr_df.dropna(how="any").reset_index(drop=True)
```

```python
hr_df.info()
```

```python
hr_df.drop("Over18", axis=1, inplace=True)
```

```python
hr_df.info()
```

```python
# save csv of cleaned data for Tableau vizzes (which keeps the object dtypes for visualization)
hr_df.to_csv("Clean_HRData.csv")
```

```python
# use list comprehension to encode Attrition and Gender columns since they have only 2 possible categorical values
hr_df['Attrition'] = [1 if x == "Yes" else 0 for x in hr_df.Attrition]
hr_df['Gender'] = [1 if x == "Male" else 0 for x in hr_df.Gender]
```

```python
attrition_cat = hr_df.dtypes[hr_df.dtypes == 'object'].index.to_list()
hr_df[attrition_cat].nunique()
```

```python
from sklearn.preprocessing import OneHotEncoder

# OneHotEncode cat data
enc = OneHotEncoder(sparse=False)

# Fit & transform the enc using the cat list
encode_df = pd.DataFrame(enc.fit_transform(hr_df[attrition_cat]))

# Add encoded vars to DF
encode_df.columns = enc.get_feature_names_out(attrition_cat)
encode_df.info()
```

```python
ML_ready_df = hr_df.merge(encode_df, how='left', left_index=True, right_index=True)
ML_ready_df = ML_ready_df.drop(attrition_cat, 1)
ML_ready_df.info()
```

```python
ML_ready_df.drop("Over18_Y", axis=1, inplace=True)
```

```python
ML_ready_df.info()
```

```python
ML_ready_df.to_csv("Clean_ML_HRdata.csv")
```
