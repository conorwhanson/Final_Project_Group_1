# Create Columns, Tables, and Populate Database 

```python
# Initialize the Sqlite engine
engine = create_engine('sqlite:///HR_Attrition.db', echo = True)
meta = MetaData()
```

```python
# Create the table structure
EmpPersonalInfo = Table(
    'EmpPersonalInfo', meta,
    Column("EmployeeID", Integer, primary_key = True),
    Column("Age", Integer),
    Column("EducationID", Integer, ForeignKey('EducationKey.EducationID')),
    Column("EducationFieldID", Integer, ForeignKey('EducationFieldKey.EducationFieldID')),
    Column("DistanceFromHome", Float),
    Column("Gender", String),
    Column("MarriageID", String, ForeignKey('MarriageKey.MarriageID')),
    Column("NumCompaniesWorked", Integer),
    Column("TotalWorkingYears", Integer),
    Column("LastUpdated", DateTime),
    extend_existing=True,
)

EmpWorkInfo = Table(
    'EmpWorkInfo', meta,
    Column("EmpWorkTableID", Integer, primary_key = True),
    Column("EmployeeID", Integer, ForeignKey('EmpPersonalInfo.EmployeeID')),
    Column("Attrition", String),
    Column("BusinessTravelID", String, ForeignKey('BusinessTravelKey.BusinessTravelID')),
    Column("DepartmentID", Integer, ForeignKey('DepartmentKey.DepartmentID')),
    Column("JobLevel", Integer),
    Column("JobRoleID", Integer, ForeignKey('JobRoleKey.JobRoleID')),
    Column("MonthlyIncome", Float),
    Column("PercentSalaryHike", Integer),
    Column("StandardHours", Integer),
    Column("StockOptionLevel", Integer),
    Column("TrainingTimesLastYear", Integer),
    Column("YearsAtCompany", Integer),
    Column("YearsSinceLastPromotion", Integer),
    Column("YearsWithCurrManager", Integer),
    Column("LastUpdated", DateTime),
    extend_existing=True,
)

EmpSurvey = Table(
    'EmpSurvey', meta,
    Column('EmpSurveyID', Integer, primary_key = True),
    Column("EmployeeID", Integer, ForeignKey('EmpPersonalInfo.EmployeeID')),
    Column("EnvironmentSatisfaction"),
    Column("JobSatisfaction", Integer),
    Column("WorkLifeBalance", Integer),
    Column("LastUpdated", DateTime),
    extend_existing=True,
);

ManagerSurvey = Table(
    'ManagerSurvey', meta,
    Column('ManagerSurveyID', Integer, primary_key = True),
    Column("EmployeeID", Integer, ForeignKey('EmpPersonalInfo.EmployeeID')),
    Column("JobInvolvement", Integer),
    Column("PerformanceRating", Integer),
    Column("LastUpdated", DateTime),
    extend_existing=True,
);

EducationKey = Table(
    'EducationKey', meta,
    Column("EducationID", Integer, primary_key = True),
    Column("Education", String),
    Column("LastUpdated", DateTime),
    extend_existing=True,
);

EducationFieldKey = Table(
    'EducationFieldKey', meta,
    Column("EducationFieldID", Integer, primary_key = True),
    Column("EducationField", String),
    Column("LastUpdated", DateTime),
    extend_existing=True,
);

DepartmentKey = Table(
    'DepartmentKey', meta,
    Column("DepartmentID", Integer, primary_key = True),
    Column("Department", String),
    Column("LastUpdated", DateTime),
    extend_existing=True,
);

JobRoleKey = Table(
    'JobRoleKey', meta,
    Column("JobRoleID", Integer, primary_key = True),
    Column("JobRole", String),
    Column("LastUpdated", DateTime),
    extend_existing=True,
);

MarriageKey = Table(
    'MarriageKey', meta,
    Column("MarriageID", Integer, primary_key = True),
    Column("MaritalStatus", String),
    Column("LastUpdated", DateTime),
    extend_existing=True,
);

BusinessTravelKey = Table(
    'BusinessTravelKey', meta,
    Column("BusinessTravelID", Integer, primary_key = True),
    Column("BusinessTravel", String),
    Column("LastUpdated", DateTime),
    extend_existing=True,
);
```

```python
# Create the db
meta.create_all(engine)
```

```python
# populate tables with data from dataframes
empPersonalInfo_df.to_sql('EmpPersonalInfo', con=engine, if_exists='append', index=False)
empWorkInfo_df.to_sql('EmpWorkInfo', con=engine, if_exists='append', index=False)
educationKey_df.to_sql('EducationKey', con=engine, if_exists='append', index=False)
educationFieldKey_df.to_sql('EducationFieldKey', con=engine, if_exists='append', index=False)
marriageKey_df.to_sql('MarriageKey_df', con=engine, if_exists='append', index=False)
businessTravelKey_df.to_sql('BusinessTravelKey', con=engine, if_exists='append', index=False)
departmentKey_df.to_sql('DepartmentKey', con=engine, if_exists='append', index=False)
jobRoleKey_df.to_sql('JobRoleKey', con=engine, if_exists='append', index=False)
empSurvey_df.to_sql('EmpSurvey', con=engine, if_exists='append', index=False)
managerSurvey_df.to_sql('ManagerSurvey', con=engine, if_exists='append', index=False)
```
